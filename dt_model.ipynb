```python
# Import libraries
import pandas as pd
from app import db, DTSchool
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing, tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import joblib
```

```python
le = preprocessing.LabelEncoder()

# Execute a query and store the results in a DataFrame
school_data = pd.read_sql_query(db.session.query(DTSchool).statement, db.session.bind)
for column_name in school_data.columns:
    if school_data[column_name].dtype == object:
        school_data[column_name] = le.fit_transform(school_data[column_name])
    else:
        pass
# school_data.dtypes
```

```python
X = school_data.drop(columns=['School', 'id'])
y = school_data['School']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
predictions = dt_model.predict(X_test)
score = accuracy_score(y_test, predictions)
score
```

```python
# Import libraries
import pandas as pd
from app import db, DTSchool
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing, tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import joblib

# Get inputed 'Course', 'Tuition_Fee', and 'Location'
input_course = input("Enter desired course: ")
input_tuition = input("Enter desired tuition fee: ")
input_location = input("Enter desired location: ")

# Execute a query and store the results in a DataFrame
school_data = pd.read_sql_query(db.session.query(DTSchool).statement, db.session.bind)

# Encode categorical variables
le = preprocessing.LabelEncoder()
for column_name in school_data.columns:
    if school_data[column_name].dtype == object:
        school_data[column_name] = le.fit_transform(school_data[column_name])
    else:
        pass

# Split data into training and testing sets
X = school_data.drop(columns=['School'])
y = school_data['School']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train decision tree model
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

# Save the trained model and label encoder to disk
joblib.dump(dt_model, 'school_recommender.joblib')
joblib.dump(le, 'label_encoder.joblib')

# Create a new DataFrame with the inputed data
school_data = pd.read_sql_query(db.session.query(DTSchool).statement, db.session.bind)
input_data = pd.DataFrame({'Course': [input_course], 'Tuition_Fee': [input_tuition], 'Location': [input_location]})
for column_name in input_data.columns:
    if input_data[column_name].dtype == object:
        input_data[column_name] = le.fit_transform(input_data[column_name])
    else:
        pass

# Use the model to make predictions on the encoded inputed data
predictions = dt_model.predict([[input_course, input_tuition, input_location]])

# Filter the recommended schools based on inputed 'Course', 'Tuition_Fee', and 'Location'
filtered_schools = school_data[(school_data['Course'] == input_course[0]) &
                               (school_data['Tuition_Fee'] == input_tuition[0]) &
                               (school_data['Location'] == input_location[0])]

# Print the filtered recommended schools
print("Recommended schools:")
print(filtered_schools)


```
