```python
import pandas as pd
import prophet as Prophet
from datetime import datetime
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
ama_bsa = pd.read_csv('CSV/board-exam-results - AMA-Accountancy.csv')
```

```python
ama_bsa['Year'] = ama_bsa['Time Date'].apply(lambda x: str(x)[-4:])
ama_bsa['Month'] = ama_bsa['Time Date'].apply(lambda x: str(x)[-6:-4])
ama_bsa['Day'] = ama_bsa['Time Date'].apply(lambda x: str(x)[:-6])
ama_bsa['ds'] = pd.DatetimeIndex(ama_bsa['Year']+'-'+ama_bsa['Month']+'-'+ama_bsa['Day'])
```

```python
ama_bsa.drop(['Time Date', 'School', 'Course', 'Year', 'Month', 'Day'], axis=1, inplace=True)
ama_bsa.columns = ['y', 'ds']
```

```python
m = Prophet.Prophet(interval_width=0.95)
m.add_seasonality(name='yearly', period=365.25, fourier_order=10)
m.add_seasonality(name='semi-annual', period=365.25/2, fourier_order=10)
ama_bsa_model = m.fit(ama_bsa)
```

```python
# Create a list of dates for which you want to make predictions
specific_dates = [datetime(2023, 6, 1), datetime(2023, 12, 29), datetime(2024, 6, 1), datetime(2024, 12, 29), datetime(2025, 6, 1), datetime(2025, 12, 29), datetime(2026, 6, 1), datetime(2026, 12, 29), datetime(2027, 6, 1), datetime(2027, 12, 29), datetime(2028, 6, 1), datetime(2028, 12, 29), datetime(2029, 6, 1), datetime(2029, 12, 29), datetime(2030, 6, 1), datetime(2030, 12, 29)]

# Create a dataframe with those specific dates
ama_bsa_future = pd.DataFrame({'ds': specific_dates})

# Get predictions for those specific dates
ama_bsa_forecast = m.predict(ama_bsa_future)
ama_bsa_forecast['yhat'] = ama_bsa_forecast['yhat'].clip(lower=0, upper=100).round()
```

```python
ama_bsa_forecast[['ds', 'yhat']]
```

```python
ama_bsa_plot = m.plot(ama_bsa_forecast)
```
